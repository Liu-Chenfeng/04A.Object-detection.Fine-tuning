````python
import os
import random
from ultralytics import YOLO
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
````

````python
# ÈÖçÁΩÆÊ®°ÂûãË∑ØÂæÑÂíåÂõæÁâáÊñá‰ª∂Â§πË∑ØÂæÑ
model_path = r"C:\Users\lcf14\Desktop\homework\Machine_Learning_appli\Fine_Tuning\YOLOv8_Fine-Tuning\weights\best.pt"
image_folder = r"C:\Users\lcf14\Desktop\homework\Machine_Learning_appli\Fine_Tuning\WatermarkDataset\sample_images_100"
save_dir = r"C:\Users\lcf14\Desktop\homework\Machine_Learning_appli\Fine_Tuning\WatermarkDataset\sample_images_100\results"

# Âä†ËΩΩËÆ≠ÁªÉÂ•ΩÁöÑ YOLO Ê®°Âûã
model = YOLO(model_path)

# Ëé∑ÂèñÊñá‰ª∂Â§π‰∏≠ÁöÑÊâÄÊúâÂõæÁâáË∑ØÂæÑ
image_files = [os.path.join(image_folder, file) for file in os.listdir(image_folder) if file.endswith(('.jpg'))]
````

````python
# ÈöèÊú∫ÈÄâÊã© 5 Âº†ÂõæÁâá
random_images = random.sample(image_files, 5)

# ÂØπÊØèÂº†ÂõæÁâáËøõË°åÊ£ÄÊµãÂπ∂‰øùÂ≠òÁªìÊûú
results = []
for img_path in random_images:
    result = model.predict(source=img_path, save=True, project=save_dir)  # Ëá™Âä®‰øùÂ≠òÊ£ÄÊµãÁªìÊûú

    # Ëé∑Âèñ‰øùÂ≠òÁªìÊûúÁöÑÊñá‰ª∂Â§πË∑ØÂæÑÔºåÂÅáËÆæ "predictX" Êñá‰ª∂Â§πÂêç‰∏∫ÁªìÊûúÊñá‰ª∂Â§π
    result_folder = os.path.join(save_dir, f"predict{len(os.listdir(save_dir))}")
    
    # Ëé∑ÂèñÁªìÊûúÂõæÂÉèÁöÑË∑ØÂæÑÔºå‰ΩøÁî®ÂéüÂßãÂõæÂÉèÁöÑÊñá‰ª∂Âêç
    detected_image_path = os.path.join(result_folder, os.path.basename(img_path))

    results.append((img_path, detected_image_path))
````

````python
# ÊòæÁ§∫Ê£ÄÊµãÁªìÊûú
for img_path, detected_image_path in results:
    # Âä†ËΩΩÂéüÂõæ
    original_image = Image.open(img_path)

    # Âä†ËΩΩÊ£ÄÊµãÂêéÁöÑÂõæÁâá
    detected_image = Image.open(detected_image_path)

    # ÊòæÁ§∫ÂõæÁâá
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(original_image)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("Detected Image")
    plt.imshow(detected_image)
    plt.axis("off")

    plt.show()
````
